### Reading the dataset

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

### Remove Duplicates

```python
df = pd.read_csv("brain_stroke.csv")

df = df.drop_duplicates()
```

### Map Categorical Values

```python
# Mapping 'gender' column
df['gender'] = df['gender'].replace({'Male': 1, 'Female': 0, 'Other': -1})

# Mapping 'ever_married' column
df['ever_married'] = df['ever_married'].map({'Yes': 1, 'No': 0})

# Mapping 'Residence_type' column
df['Residence_type'] = df['Residence_type'].map({'Urban': 1, 'Rural': 0})

# Correcting and Mapping 'work_type'
df['work_type'] = df['work_type'].replace({'Govtjov': 'Govt_job'})

# One-hot encoding for 'smoking_status' for clarity in different statuses
df = pd.get_dummies(df, columns=['smoking_status'], drop_first=True)
```

```python
# Check if 'work_type' is correctly mapped, if necessary
work_type_mapping = {
    'Private': 0,
    'Self-employed': 1,
    'Govt_job': 2,
    'children': 3,
    'Never_worked': 4
}
# Assuming work_type was supposed to be categorical, ensure it's correctly mapped
df['work_type'] = df['work_type'].replace(work_type_mapping)

# Verify the data types after mapping
print(df.dtypes)
```

```python
df.fillna(0, inplace=True)

# Verify changes
print(df.info())
print(df.head())
```

# Exploratory Data Analysis

```python
# Check the counts of each unique value in categorical columns to ensure encoding worked as expected

print("Gender Distribution:")
print(f"Female: {df['gender'].value_counts()['Female']}")
print(f"Male: {df['gender'].value_counts()['Male']}")

print("\nWork Type Distribution (Encoded):")
print(f"Private: {df['work_type'].value_counts()[0]}")
print(f"Self-employed: {df['work_type'].value_counts()[1]}")
print(f"Govt_job: {df['work_type'].value_counts()[3]}")
print(f"Children: {df['work_type'].value_counts()[2]}")

print("\nEver Married Distribution:")
print(f"Yes: {df['ever_married'].value_counts()['Yes']}")
print(f"No: {df['ever_married'].value_counts()['No']}")

print("\nResidence Type Distribution:")
print(f"Urban: {df['Residence_type'].value_counts()['Urban']}")
print(f"Rural: {df['Residence_type'].value_counts()['Rural']}")

print("\nStroke Distribution:")
print(f"No Stroke: {df['stroke'].value_counts()[0]}")
print(f"Stroke: {df['stroke'].value_counts()[1]}")
```

## Visualize Distributions

### Age Distributions

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### BMI and Average Glucose Level Distributions

```python
# Plot BMI distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['bmi'], bins=30, kde=True, color='purple')
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()

# Plot Average Glucose Level distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['avg_glucose_level'], bins=30, kde=True, color='orange')
plt.title('Average Glucose Level Distribution')
plt.xlabel('Average Glucose Level')
plt.ylabel('Frequency')
plt.show()
```

### Stoke count by gender, residence type,smoking status

```python
# Stroke by gender
plt.figure(figsize=(8, 5))
sns.countplot(x='gender', hue='stroke', data=df)
plt.title('Stroke Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Stroke')
plt.show()

# Stroke by Residence Type
plt.figure(figsize=(8, 5))
sns.countplot(x='Residence_type', hue='stroke', data=df)
plt.title('Stroke Count by Residence Type')
plt.xlabel('Residence Type')
plt.ylabel('Count')
plt.legend(title='Stroke')
plt.show()


# Plotting Stroke Count by Smoking Status
plt.figure(figsize=(10, 6))
sns.countplot(x='smoking_status', hue='stroke', data=df)
plt.title('Stroke Count by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.show()
```

### ANALYSIS

From this correlation heatmap, we can interpret the relationships between various features in the dataset. Here are some key observations:

1. **Age and Ever Married**: There is a strong positive correlation (0.68) between age and the "ever married" variable, suggesting that older people are more likely to have been married.

2. **Age and BMI**: Age also has a moderate positive correlation with BMI (0.37), which could indicate that BMI tends to increase with age in this dataset.

3. **Hypertension and Age**: There's a moderate positive correlation (0.28) between age and hypertension, indicating that hypertension is more common in older individuals.

4. **Stroke and Age**: Stroke has a mild positive correlation (0.25) with age, implying that strokes are more common among older people.

5. **Smoking Status**:
   - Former smokers have a slight positive correlation with age (0.24), possibly indicating that older people are more likely to be former smokers.
   - Current smokers have a slight negative correlation with those who never smoked (-0.35).

6. **Stroke and Health Conditions**:
   - Stroke shows slight positive correlations with hypertension (0.13) and heart disease (0.13), suggesting that individuals with these conditions may have a higher likelihood of experiencing a stroke.

7. **Avg Glucose Level and Health Conditions**: The average glucose level has moderate correlations with both heart disease (0.17) and hypertension (0.17), indicating a potential association with these health conditions.
